**Python** is an **interpreted**, high-level, general-purpose programming language characterized by **strong** and **dynamic typing**, as well as automatic memory management.
It supports multiple programming paradigms, including procedural, functional, and object-oriented programming.

**Dynamic typing** means a variable can hold values of **different types** during program execution.

```python
# assigning an integer value to a variable
x = 5

# checking the type of the variable
print(type(x))

# assigning a string value to the same variable
x = "and now for sth completely different"

# checking the type again
print(type(x))

# assigning a floating-point value to the same variable
x = 5.0

# checking the type again
print(type(x))

# printing the current value of the variable
print(x)
```

**Strong typing** means that Python does not perform implicit type conversions. Let’s create our first error!

```python
x = 5
y = "5"

print(x + y)
```

However, this operation is allowed for the numeric types **float** and **int**; the result will be of type float.

```python
x = 5
y = 5.0

print(x + y)
```

Value equality is compared using the `==` operator.

```python
x = 100000
y = 100000

print(id(x))
print(id(y))

print(x == y)
print(x is y)
```

#### Basic Operations on Numeric Types

```python
# integer (whole number variable)
i = 255

# alternative notations
i = 0xff        # hexadecimal
i = 0o377       # octal
i = 0b11111111  # binary

# float (floating-point variable)
f = 0.25

# exponential notation
f = 2.5e-01

# complex (complex numbers)
c = 3.5 + 0.5j

# undefined / null-like value
x = None

# mathematical operations
y = i + 1   # addition
y = i - 1   # subtraction
y = i * 1   # multiplication
y = i / 1   # division
y = i % 5   # modulo (remainder)
y = i // 5  # floor division
y = i ** 3  # exponentiation

y = -i      # negation
y = abs(i)  # absolute value

# augmented assignment
i += 1      # short form of i = i + 1
i -= 1      # short form of i = i - 1
i *= 2      # short form of i = i * 2
i **= 3     # short form of i = i ** 3 (exponentiation)

i = int(i)      # type conversion
```

#### Boolean Values and Operations

```python
# bool (boolean, logical values)
# True (1) or False (0)
# bool is a subclass of int
# Logical expressions evaluate to True or False.

b = True

# comparison operators
b = i <= 255
b = i == 255
b = i != 255

# logical operators
b = i < 0 and i > 255
b = i < 0 or i < 255
b = (i >= 0 and i < 256) or i % 10 == 5
b = not (i >= 0 and i < 10) or i % 10 == 5

# identity
b = x is None
b = x is not None

# membership
b = 10 in [0, 10, 20, 30]
b = 10 not in [0, 1, 2, 3]

# any() and all()
b = any([i == 0, i == 255, i == 300])
b = all([i > 0, i == 255, i < 256])
```

##### Mutability and References

```python
# mutability

l1 = ['a', 'b', 'c']
l2 = l1      # l2 refers to the same list object

l2[1] = 'this was a reference'

print(l1)
print(l1 == l2)   # same value
print(l1 is l2)   # same object
```

Both variables refer to the same list object, so modifying `l2` also changes `l1`.

```python
l1 = ['a', 'b', 'c']

l2 = l1.copy()    # or: l2 = l1[:]
l2[1] = 'this is a different object'

print(l1)
print(l2)
print(l1 == l2)   # values may differ
print(l1 is l2)   # different objects
```

`copy()` or slicing `[:]` creates a new list object.

Beware shallow copies.

```python
l1 = [[1,2],[3,4]]
l2 = l1.copy()

l2[0][0] = 99
print(l1)
```

We could use `deepcopy()`, which recursively copies all nested objects, but it is generally better to create a new list (or data structure) explicitly when needed.

```python
import copy

l1 = [[1, 2], [3, 4]]
l2 = copy.deepcopy(l1)

l2[0][0] = 99

print(l1)
print(l2)
```

#### Unordered Collections

```python
# A set is an unordered collection of unique, hashable elements.

s = {1.0, 'text', 4.235}

# add a single element
s.add(2)

# add elements from another collection
s.update(['a', 'b', 'c'])

# remove a single element
s.discard('text')

print(s)
```

```python
# convert a list/tuple to a set
l = [0, 1, 1, 1, 1, 1, 2, 3]
s = set(l)

print(s)
```

```python
# frozenset — immutable version of a set
fs = frozenset(l)

print(s, fs, sep='\n')
```

##### Set Operations

```python
s1 = {1, 2, 3, 4, 5, 6}
s2 = {4, 5, 6, 7, 8, 9}

# union — elements from s1 or s2
union = s1 | s2
union = s1.union(s2)

# difference — elements in s1 not in s2
diff = s1 - s2
diff = s1.difference(s2)

# intersection — elements common to both sets
inter = s1 & s2
inter = s1.intersection(s2)

# symmetric difference — elements unique to each set
sym = s1 ^ s2
sym = s1.symmetric_difference(s2)
```

#### Dictionaries

```python
# dict (dictionary) is a collection
# of key : value pairs
# keys must be hashable (e.g. numbers, strings, tuples)

d = dict()
d = {}

d = {
    '1': 'a',
    2.0: ['b', 'b', 'b'],
    (1, 2): 'c',
    10: 20.5
}

print(d['1'])
print(d[(1, 2)])
```

```python
# add a new key-value pair
d['new key'] = 'new value'

print(d)
```

```python
# or using the update method (it can be used to change multiple key-value pairs at the same time)

d = {'x': 10}
d.update({'y': 20, 'x' : 1})
print(d)
```

```python
# safe access using get()
# get() returns the value for a key or a default value if the key is missing.

print(d.get('1'))
print(d.get('missing'))          # returns None
print(d.get('missing', 'default'))  # custom default value
```

```python
# Membership checks dictionary keys, not values.
print('1' in d)
```

```python
# Removing elements
d.pop('1')
print('1' in d)
```

```python
# dictionary views

print(d.keys())    # keys
print(d.values())  # values
print(d.items())   # key-value pairs
```

### Running .py Scripts from the Command Line (Linux)

A script should begin with a shebang (#!) followed by the path to the interpreter.
The Unix utility env searches for the interpreter in the $PATH environment variable, which improves script portability.
