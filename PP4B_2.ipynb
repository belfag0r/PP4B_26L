## Control Flow

The order in which instructions are executed may differ from their order in the source code.

The most common control flow mechanisms are:

- **selection** — conditional execution (`if`, `elif`, `else`)
- **loops** — repeated execution (`for`, `while`)
- **jumps** — transferring execution to another point in the program (`break`, `continue`, `return`)*

*these are not actual jump statements like `goto` in C/C++

#### Conditional statements (if, elif, else)

Conditional statements allow a program to execute code depending on whether a condition is true or false.

```python
x = -1

# this condition is always evaluated first
if x == 0:
    print("x is equal to zero")

# elif (else if)
# this condition is checked only if the preceding **if** is False
elif x > 0:
    print("x is positive")

# this condition is checked only if the previous **elif** is False
elif x < 0:
    print("x is negative")

# this block executes only if none of the above conditions are True
else:
    print("x is something else")
```

##### Note
In longer conditional chains, it is often beneficial to check the most likely condition first, so the program can terminate the chain earlier and avoid unnecessary evaluations.

#### Loops — `for` and `while` statements

Loops allow a program to execute a block of code repeatedly.

Python provides two basic looping constructs:

* `for` — iterates over elements of an iterable (e.g., list, string, range)
* `while` — repeats execution as long as a condition is True

```python
# basic for loop
collection = list(range(5))

for element in collection:
    print(element, end=" ")
```

```python
# nested for loop
for i in range(0, 4, 1):
    for j in range(4):
        print(j, end=" ")
        
    print("")
```

```python
# basic while loop

x = 1
while x < 10:
    print(x, end=" ")
    x += 1
```

##### Iterating over dictionaries

Dictionaries can be iterated in several ways, depending on whether we want to access keys, values, or key–value pairs.

```python
example_dict = {"gene": "BRCA1", "species": "Homo sapiens", "length": 1863}

# iterate over keys
for key in example_dict:
    print(key, end=", ")
```

```python
# iterate over values
for value in example_dict.values():
    print(value, end=", ")
```

```python
# iterate over key–value pairs
for key, value in example_dict.items():
    print(f'{key} : {value}')
```

##### `enumerate()` — iterating with indices

When iterating over a sequence, we sometimes need both the element and its index.
A common beginner pattern is:

```python
sequence = ["A", "T", "G", "C"]

for i in range(len(sequence)):
    print(i, sequence[i])
```

While this works, it is generally not recommended, because it is less readable and requires manual indexing.

A more Pythonic approach is to use `enumerate()`:
* enumerate() returns pairs of (index, element)
* These pairs can be unpacked directly in the loop
* The starting index can be specified using the start argument

```python
sequence = ["A", "T", "G", "C"]

for i, base in enumerate(sequence):
    print(i, base)
```

##### Iterating over nested data structures

In practice, data structures can be deeply nested, combining dictionaries and lists. Accessing the desired values may require multiple levels of iteration and indexing.

```python
d = {
    "K9O3": [{"a": [1, 2, 3]}, {"b": [{"a": 1}]}],
    "D2F8": [{"a": [3, 2, 1]}, {"b": [{"a": 2}]}],
    "B6J9": [{"a": [1, 1, 1]}, {"b": [{"a": 3}]}],
    "1B0D": [{"a": [0, 0, 0]}, {"b": [{"a": 4}]}],
}

for some_id in d:
    for entity in d[some_id]:

        if "b" in entity:

            print(some_id, entity["b"][0]["a"])
```

#### `break` and `continue`

The statements break and continue modify the normal execution of a loop.

* `break` immediately terminates the loop.

* `continue` skips the remaining statements in the current iteration and proceeds to the next iteration.

```python
from random import randint

while True:                            # while True creates an infinite loop

    n = randint(0, 5)                  # In each iteration, a random number between 0 and 5 is generated

    if n == 5:                         # If the number equals 5, the message is printed and break exits the loop
        print(n, "is 5!")
        break

    else:                              # Otherwise, continue starts the next iteration
        print(n, "is not 5!")
        continue                      

    print(":'(")                      # The final print(":'(") is never executed, because 
                                      # the loop either continues earlier or terminates with break
```

## Functions

#### Defining a function
A new function is defined using the keyword `def`.
By convention (snake_case, see PEP 8*), function names use lowercase letters, with words separated by underscores `_`.

Python emphasizes code readability, as code is typically read more often than it is written. Well-chosen names for variables and functions often make additional comments unnecessary.

After the function name, parentheses `()` specify parameters. Parameters may include default values.

The definition line ends with `:`, followed by an indented code block executed when the function is called.

A function may include the keyword `return`, followed by a value or expression to return. If no return statement is provided, the function returns `None`.

<br>
<br>

*PEP 8 – Style Guide for Python Code

https://peps.python.org/pep-0008/#function-and-variable-names

```python
def add_numbers(x, y=1): # x and y are parameters of the function; y=1 defines a default value
    """Return the sum of x and y."""
    result = x + y
    return result

print(add_numbers(4))

# The function definition contains a docstring.
# It can be accessed using the special attribute __doc__.
print(add_numbers.__doc__)
```

```python
# The built-in help() function also displays the docstring.
help(add_numbers)
```

The keyword `pass` is used when Python requires a statement syntactically, but no action should be performed. It acts as a placeholder, allowing the program to run without executing any operation.

This is commonly used when defining functions, classes, or control structures that will be implemented later.

```python
def sorry_i_do_not_have_an_idea_yet():
    pass
```

#### Variable-Length Arguments (*args, **kwargs)

Functions can also accept a variable number of arguments using `*args` and `**kwargs`.

* `*args` collects additional positional arguments into a **tuple**.

* `**kwargs` collects additional keyword arguments into a **dictionary**.

```python
# *args – positional (non-keyword) arguments are collected into a tuple

def input_a_bunch_of_stuff(*args):
    print(type(args))
    print(sum(args))
    print(*args)

input_a_bunch_of_stuff(3, 4, 1)
```

```python
# **kwargs – keyword arguments are collected into a dictionary

def input_a_bunch_of_stuff(**kwargs):

    print(type(kwargs))
    print(kwargs)
    print(*kwargs)

    if 'z' in kwargs:
        print(f"value of 'z': {kwargs['z']}")

input_a_bunch_of_stuff(x=1, y=1, z=1)
```

Positional arguments must appear before keyword arguments in a function call.

```python
# extreme example

def input_a_lot_of_stuff(a, b=5, *args, required_kw_argument, **kwargs):

    print("args =", args)
    print(f"a={a}, b={b}")
    print("kwargs =", kwargs)

# b has a default value but is overridden
input_a_lot_of_stuff(9, 9, 1, 1, 1, 1, 1, required_kw_argument=1, x=1, y=1, z=1)
```

This function demonstrates several types of parameters:
* `a` — required positional argument

* `b=5` — positional argument with a default value

* `*args` — collects additional positional arguments into a tuple

* `required_kw_argument` — a required keyword-only argument

* `**kwargs` — collects additional keyword arguments into a dictionary

In the example call:

* `b` had a default value (5), but it is overridden by the argument 9.

* The extra positional arguments are stored in args.

* Additional keyword arguments (x, y, z) are stored in kwargs.

#### Multiple return statements

A function may contain multiple `return` statements, allowing it to exit early when a condition is met.

```python
def is_my_number_good_check(my_number):

    if my_number < 15:
        return "No"

    elif 15 < my_number <= 22:
        return "Yes"

    else:
        return "No"

is_my_number_good_check(30) # Each return immediately terminates the function and returns a value
```

```python
def is_my_number_good_check(my_number):

    if 15 < my_number <= 22:
        return "Yes"

    return "No"
```

##### Note on style

Using multiple return statements is common and accepted in Python.
They often improve readability by avoiding deeply nested if statements.

However, in some programming styles (especially in older languages), developers prefer a single return point for easier control flow analysis.

##### Returning multiple values

```python
def sequence_stats(sequence):
    length = len(sequence)
    gc = (sequence.count("G") + sequence.count("C")) / length
    return length, gc # Functions can return multiple values, which are returned as a tuple

# the returned tuple can be unpacked like this
length, gc = sequence_stats("ATGCGC")

# alternatively, store the result and access elements by index
result = sequence_stats("ATGCGC")
length = result[0]
gc = result[1]
```

#### Local and Global Namespaces

Functions introduce the concept of a **namespace**, which is fundamental in Python.
As stated in The Zen of Python:

<p style="text-align:center;">"Namespaces are one honking great idea — let's do more of those!"</p>

A namespace is a mapping between names (identifiers) and objects.

Variable scope (LEGB rule)

Python resolves variable names according to the LEGB hierarchy:

    Built-in – built-in names and functions (e.g., print, len)

        Global – names defined at the module (file) level

            Enclosing – names in the scope of enclosing functions
                        (in nested functions)

                Local – names defined inside the current function

When Python encounters a variable name, it searches for it in this order:
<p style="text-align:center;">Local → Enclosing → Global → Built-in</p>

When a function is called, it creates its own local namespace for the duration of its execution.

Variables defined inside the function exist only in this local scope and are not accessible from outside the function. After the function finishes, its local namespace is discarded.

```python
x = "global variable"

def function():

    y = "local variable"
    print(x)

function()

print(y)  # at this point y no longer exists
```

When Python interprets code, it searches for a variable starting from the lowest scope. If the name is not found there, Python continues searching in the next higher scope. This process follows the **LEGB rule**.

```python
x = 1

def outer():

    x = 2

    def inner():

        x = 3
        print("In inner x =", x)

    inner()
    print("In outer x =", x)

outer()
print("Global variable x =", x)
```

The execution of Python code can be visualized step by step using https://pythontutor.com/visualize.html

```python
x = 1

def outer():

    x = 2

    def inner():

        # x = 3
        print("In inner x =", x)

    inner()
    print("In outer x =", x)

outer()
print("Global variable x =", x)
```

```python
x = 1

def outer():

    # x = 2

    def inner():

        # x = 3
        print("In inner x =", x)

    inner()
    print("In outer x =", x)

outer()
print("Global variable x =", x)
```

##### `global` and `nonlocal`

The keywords `global` and `nonlocal` allow us to modify variables defined in outer scopes.

* global refers to variables in the global (module) scope.

* nonlocal refers to variables in the nearest enclosing function scope.

These keywords change the scope in which a variable is looked up and modified.

However, modifying outer-scope variables in this way is generally discouraged, as it can make code harder to understand and maintain. In most cases, it is preferable to pass values as function arguments and return results explicitly.

```python
x = 1

def outer():

    x = 2

    def inner():

        global x
        x = 3
        print("In inner x =", x)

    inner()
    print("In outer x =", x)

outer()
print("Global variable x =", x)
```

```python
x = 1

def outer():

    x = 2

    def inner():

        nonlocal x
        x = 3
        print("In inner x =", x)

    inner()
    print("In outer x =", x)

outer()
print("Global variable x =", x)
```

#### Type hints
Python supports optional type hints, which allow you to indicate the expected types of function arguments and return values. Type hints do not change how the program runs.

##### Without type hints

```python
def count_bases(sequence):
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "C": sequence.count("C"),
        "G": sequence.count("G"),
    }
```

##### With basic type hints

```python
def count_bases(sequence: str) -> dict:
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "C": sequence.count("C"),
        "G": sequence.count("G"),
    }
```

##### With more specific typing (Python ≥ 3.9)

```python
def count_bases(sequence: str) -> dict[str, int]:
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "C": sequence.count("C"),
        "G": sequence.count("G"),
    }
```

##### Using the `typing` module (older style)

```python
from typing import Dict

def count_bases(sequence: str) -> Dict[str, int]:
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "C": sequence.count("C"),
        "G": sequence.count("G"),
    }
```

Using type hints makes sense in larger collaborative projects or during library development, but they are often unnecessary in small scripts or exploratory data analysis.
